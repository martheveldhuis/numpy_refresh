For nd arrays, we index the nth dim with a comma, which is different from python's notation

```python
ary[0, 1]
```

```python
ary[-1, 1]
```

```python
ary[:, 0] # first column
```

```python
ary[::-1]
```

```python
ary[0] # first row
```

```python
ary[:, :2] # first two columns
```
