```python
def set_no_match(odds, girl, guy):
    odds.at[guy, girl] = 0
    
    guy_zeros = odds.loc[guy].loc[odds.loc[guy].isin([0]) == True]
    girl_zeros = odds[girl].loc[odds[girl].isin([0]) == True]
    
    guy_zeros_count = guy_zeros.count()
    girl_zeros_count = girl_zeros.count()
    
    # reduce denominator by 1 for each zero counted
    denominator_guy = odds.shape[0] - guy_zeros_count
    denominator_girl = odds.shape[1] - girl_zeros_count
    
    guys_to_update = set(odds[girl].index) - set(girl_zeros.index)
    girls_to_update = set(odds.loc[guy].index) - set(guy_zeros.index)
    
    odds[girl].loc[guys_to_update] = 1 / denominator_girl
    odds.loc[guy].loc[girls_to_update] = 1 / denominator_guy
    
    return odds    
```
